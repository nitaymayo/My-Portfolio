# Circuits to show and implement Quantum Teleportation

Teleporting the state of Qubit 0 to Qubit 2

```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
# Start building the circuit
qc = QuantumCircuit(3,3)

qc.x(0)

qc.barrier()
# Creating Entangelment
qc.h(1)
qc.cx(1,2)

qc.barrier()

qc.draw()
```

```python
qc.cx(0, 1)

qc.h(0)

qc.barrier()

qc.measure(0,0)
qc.measure(1,1)

qc.barrier()

qc.draw()
```

```python
# Finishing the circuit and plotting
qc.cx(1,2)
qc.cz(0,2)

qc.measure(2,2)
qc.draw()
```

```python
Aer.backends()
```

```python
# Executing and Demonstrating the teleportation
# Note that the measured Third Qubit state is the same as the initial First Qubit state, hanse the teleportation
simulator = Aer.get_backend('qasm_simulator')

job = execute(qc, backend=simulator)

plot_histogram(job.result().get_counts(qc))
```
